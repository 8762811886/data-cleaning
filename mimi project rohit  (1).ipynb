###### EDA

```python
#Checking for unique categories in apps
# there are 8268 free and 610 paid apps
# we can visualise this too but here we are not going to do so
df["Type"].value_counts() 
```

```python
#  since we know that reviews data type is float, so we will try to change it 
# so that we can apply proper operations easily
df.Reviews= df.Reviews.astype(float) 
```

```python
# now
df.describe()
```

```python
# we know that installs column has + sign at the end of them
# we need to remove that, so that we can work on that column easier

df.Installs= df["Installs"].str.replace("+", "")
```

```python
df['Installs'].head()
```

```python
# there were float values that have comas inbetween then
# so we are going to remove those commas for easy working
df.Installs= df["Installs"].str.replace(",", "")
```

```python
df["Installs"] = pd.to_numeric(df["Installs"])
```

```python
# now we are going replace the dollar signs, that are present in price column 
df["Price"]= df["Price"].str.replace("$", "")
```

```python
# now converting data of price to float
df["Price"]= df.Price.astype(float)
```

```python
df["Price"].head()
```

```python
df.describe()
# Apply different operations to these columns would be easier
```

```python
#now Correlation Heatmap
plt.figure(figsize=(10,5))
corr= df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# checking for unique values, as we already know that there are varying sizes of apps 
# but it can also be observed that on the atail of every number there is either mb or k written
# we need to clear that so that we can work on that column properly
df.Size.unique()
```

```python
#  removing the "m" which is the mb for the size

df.Size= df["Size"].str.replace("M", "")
df.Size= df['Size'].str.replace("Varies with device","-1")
```

```python
#Here we replace k and change the unit to Mb
df['Size']=df['Size'].apply(lambda x: str(round((float(x.rstrip('k'))/1024),1) )if x[-1]=='k' else x)
```

```python
df.Size.unique()
```

```python
df.drop(['Genres','Last Updated','Current Ver',"Android Ver"],axis=1,inplace=True)
# we know that genres and category are same, so we drop that
# rest last updare, currentver, android ver are not relevant to us 
```

```python
# figure related code
fig = plt.figure()
fig.suptitle('Boxplot', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
ax.boxplot(df['Rating'])
ax.set_title('Rating class v/s rating')
ax.set_xlabel('Rating class')
ax.set_ylabel('Rating')

plt.show()
```

```python
# Count plot for rating
plt.figure(figsize=(15,10))
sns.set_theme(style="darkgrid")
#dat = sns.load_dataset("df")
plt.xticks(rotation=90)
plt.title("Rating V/s Count")
ax = sns.countplot(x="Rating", data=df)
```

```python
#We look at the top five value counts  for rating 

df["Rating"].value_counts().nlargest(5).sort_values(ascending=True).plot.bar()
plt.ylabel("Count")
plt.xlabel("Rating")
plt.title("Playstore - Top 5 App Rating count")
plt.xticks(rotation=45)

plt.show()
```
